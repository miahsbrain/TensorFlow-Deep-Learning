```python
image_shape = (130, 130, 3)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
help(ImageDataGenerator)
```

```python
image_gen = tf.keras.preprocessing.image.ImageDataGenerator(rotation_range=20, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.1, zoom_range=0.1, horizontal_flip=True, fill_mode='nearest')
```

```python
para_image = imread(para_cell)
plt.imshow(para_image)
```

```python
plt.imshow(image_gen.random_transform(para_image))
```

```python
image_gen.flow_from_directory(train_path)
```

```python
image_gen.flow_from_directory(test_path)
```

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=(3,3), input_shape=image_shape, activation='relu'))
model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))

model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), input_shape=image_shape, activation='relu'))
model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))

model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), input_shape=image_shape, activation='relu'))
model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))

model.add(tf.keras.layers.Flatten())

model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dropout(0.5))

model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
callbacks = tf.keras.callbacks
```

```python
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=2, verbose=1)
```

```python
batch_size = 16
```

```python
train_image_generator = image_gen.flow_from_directory(train_path, target_size=image_shape[:2], color_mode='rgb', batch_size=batch_size, class_mode='binary')
test_image_generator = image_gen.flow_from_directory(test_path, target_size=image_shape[:2], color_mode='rgb', batch_size=batch_size, class_mode='binary', shuffle=False)
```

```python
train_image_generator.class_indices
```

```python
results = model.fit_generator(train_image_generator, validation_data=test_image_generator, epochs=20, callbacks=[early_stopping])
```

```python
metrics = pd.DataFrame(model.history.history)
```

```python
metrics[['loss', 'val_loss']].plot()
```

```python
metrics[['accuracy', 'val_accuracy']].plot()
```

```python
pred = model.predict(test_image_generator)
```

```python
predictions = (pred > 0.5).astype('int32')
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_test
```

```python
print(sk.metrics.classification_report(y_test, predictions))
```
