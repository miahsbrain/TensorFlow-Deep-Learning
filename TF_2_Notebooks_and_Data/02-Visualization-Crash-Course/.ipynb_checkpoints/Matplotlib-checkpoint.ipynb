```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [100,200,300,300,100,200]
```

```python
plt.plot(x, y)
plt.show()
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                       'price':[100,120,190,200,230,310,330,305]})
```

```python
plt.plot(x, y, color='orange', marker='o', markersize=3, linestyle='--')

plt.xlim(min(x), max(x)+min(x))
plt.ylim(min(y), max(y)+min(y))

plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```
