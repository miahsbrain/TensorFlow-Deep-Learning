```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
iris = pd.read_csv('../../DATA/iris.csv')
```

```python
iris.head()
```

```python
X = iris.drop('species', axis=1)
```

```python
y = iris['species']
```

```python
y.unique()
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
encoder = LabelBinarizer()
```

```python
y = encoder.fit_transform(y)
```

```python
y
```

```python
encoder.classes_
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
scaler = MinMaxScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
layers = tf.keras.layers
models = tf.keras.models
```

```python
model = models.Sequential([
    layers.Dense(4, activation='relu', input_shape=[4]),
    layers.Dense(3, activation='softmax')
])
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
callbacks = tf.keras.callbacks
```

```python
early_stop = callbacks.EarlyStopping(patience=10, verbose=1)
```

```python
model.fit(scaled_X_train, y_train, validation_data=(scaled_X_test, y_test),epochs=300, callbacks=[early_stop])
```

```python
metrics = pd.DataFrame(model.history.history)
```

```python
metrics
```

```python
metrics[['loss', 'val_loss']].plot()
```

```python
metrics[['accuracy', 'val_accuracy']].plot()
```

```python
model.evaluate(scaled_X_test, y_test)
```

```python
epochs = len(metrics)
```

```python
epochs
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
model = models.Sequential([
    layers.Dense(4, activation='relu', input_shape=[4]),
    layers.Dense(3, activation='softmax')
])
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(scaled_X, y, epochs=epochs)
```

```python
model.save('final_iris_model.keras')
```

# Prediction Function

```python
iris.head(1)
```

```python
flower_example = {'sepal_length': 5.1, 'sepal_width':3.5, 'petal_length':1.4, 'petal_width': 0.2}
```

```python
def return_prediction(model, scaler, sample_json):
    s_len = sample_json['sepal_length']
    s_wid = sample_json['sepal_width']
    p_len = sample_json['petal_length']
    p_wid = sample_json['petal_width']

    flower = [[s_len, s_wid, p_len, p_wid]]
    classes = np.array(['setosa', 'versicolor', 'virginica'])
    flower = scaler.transform(flower)

    class_ind = np.argmax((model.predict(flower) > 0.5).astype('int32'), axis=1)[0]

    return classes[class_ind]
```

```python
return_prediction(flower_model, flower_scaler, flower_example)
```
